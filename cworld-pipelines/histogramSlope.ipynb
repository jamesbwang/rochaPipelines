```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "allClusters.clusters"
```

```python
df = pd.read_csv(filename, sep="\t", names=["chr", "ris", "sbin", "ebin", "is"], usecols=["ris"])
```

```python
hist=np.histogram(df["ris"], bins=200)
```

```python
g = np.zeros(200)
for i in range(1,199):
    g[i] = (hist[0][i+1] - hist[0][i-1])/2
```

```python
g[0] = (hist[0][1] - hist[0][0])/2
```

```python
g[199] = (hist[0][199] - hist[0][198])/2
```

```python
plt.plot(hist[1][:-1], g)
plt.grid(b=True)
```

```python
h = np.zeros(200)
for i in range(1,199):
    h[i] = (g[i+1] - g[i-1])/2
g[0] = (g[1] - g[0])/2
h[199] = (g[199] - g[198])/2
```

```python
plt.plot(hist[1][:-1], h)
plt.grid(b=True)
```
